**Текстовый корпус** - это совокупность текстов, подобранных и обработанных по определенным правилам, которая используется для исследования языка.

### <span style="color: orange">Spell checkers</span> проверяют слова на наличие ошибок

 Компьютерная лингвистика в широком смысле занимается всеми ситуациями, когда компьютеру нужно работать с естественным языком.

#### <span style="color: orange">Задачи компьютерной лингвистики</span>
* исправление ошибок
* машинный перевод
* информационный поиск
* анализ тональности (<span style="color: orange">Sentiment analysis</span>)
* извлечение информции из текста (краткий ответ на вопрос в тексте. Важно понимать синтаксическую структуру)
* создание диалоговых агентов и голосовых помощников

#### <span style="color: orange">Дело в том, что под словосочетанием «компьютерная лингвистика» скрываются две связанные, но немного различные сферы деятельности:</span>

- автоматическая обработка естественного языка (NLP = Natural Language Processing по-английски; не путать с нейролингвистическим программированием);
- компьютерная лингвистика как фундаментальная наука, которая использует компьютеры не для того, чтобы решать прикладные задачи, а для того, чтобы узнать больше о человеческом языке.
