```python
def function_1_solution(x: torch.tensor) -> torch.tensor:
    return ((x.sin() - x.cos()) / (x ** 2 + 1)).sum()
```

# Task 3



```python
def function_2_solution(x: torch.tensor, y: torch.tensor) -> torch.tensor:
    return torch.cat(tensors=(torch.tensor([i / y.sum() for i in x]), torch.tensor([j / x.sum() for j in y])))
```

```python
x = torch.tensor(data=[1, 2])
y = torch.tensor(data=[3, 4])

print(function_2_solution(x, y))
```

# Task 3



```python
def function_3_solution(x: torch.tensor, y: torch.tensor) -> torch.tensor:
    return torch.tensor(data=[[i.log() * j ** 2 for j in y] for i in x])
```

```python
print(function_3_solution(x, y))
```

```python
def function_3_solution(x: torch.tensor, y: torch.tensor) -> torch.tensor:
    return torch.tensor([i.log() for i in x], dtype=torch.float32).reshape(-1, 1) @ torch.tensor([j ** 2 for j in y], dtype=torch.float32).reshape(1, -1)
```

```python
print(function_3_solution(x, y))
```
