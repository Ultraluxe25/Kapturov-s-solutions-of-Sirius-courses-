# Task 1



```python
def accuracy_solution(real, predicted):
    observations = len(real)
    predicted_correctly = 0
    
    for i in range(observations):
        if real[i] == predicted[i]:
            predicted_correctly += 1
            
    return predicted_correctly / observations
```

# Task 2



```python
def remove_adjacent_solution(l):
    pointer = 0
    
    while pointer < len(l) - 1:
        if l[pointer] == l[pointer + 1]:
            l.pop(pointer + 1)
        else:
            pointer += 1
            
    return l
```

```python
numbers = [5, 5, 1, 5, 5, 7, 7, 7]
print(remove_adjacent_solution(numbers))
```

## Task 3.1



```python
def string_to_words_solution(s):
    words = []
    word = ""
    
    for char in s:
        if char != " ":
            word += char
        else:
            words.append(word)
            word = ""
    else:
        words.append(word)
            
    return words
```

# Task 3.2



```python
def count_word_solution(s, word):
    words = string_to_words_solution(s)
    return words.count(word)
```
